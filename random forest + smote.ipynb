```python
from collections import Counter
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.combine import SMOTEENN 
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import roc_curve,auc
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv('/content/drive/MyDrive/mus skripsi/pc.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
y = data.heartpred
X = data.drop('heartpred', axis=1)
```

```python
sm = SMOTE (random_state=0)
X_res, y_res = sm.fit_resample(X, y)
print('Original dataset shape is ', Counter(y))
print('Resample dataset shape is ', Counter(y_res))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size = 0.3, random_state =10)
```

```python
rfc = RandomForestClassifier(n_estimators=100, random_state=3)
```

```python
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)
```

```python
from sklearn import metrics
from pandas import DataFrame
import numpy as np
```

```python
akurasirfs = metrics.accuracy_score(y_test, y_pred)
RFS = np.around(akurasirfs * 100, decimals=2)
RFS
```
